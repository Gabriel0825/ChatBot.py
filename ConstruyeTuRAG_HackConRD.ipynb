## Concepto 1 — Que es RAG y por que existe

---

Imagina que contratas a un analista nuevo en el banco.

El primer dia lo sientas en su escritorio.  
El tiene su conocimiento general — sabe de finanzas, sabe de fraude,  
estudio en la universidad. Pero no sabe nada de **tu banco especifico**.

Si le preguntas "cuantos fraudes tuvimos este mes", no sabe.  
Podria inventarlo. Podria decir un numero que suena razonable.  
Pero no tendria ninguna base real.

**Eso es un modelo de IA sin RAG.** Mucho conocimiento general. Cero datos tuyos.

---

Ahora imagina que antes de preguntarle,  
le pones en el escritorio **el reporte mensual del banco**.

Ahora si le preguntas "cuantos fraudes tuvimos este mes",  
el analista lee el reporte y te da el numero exacto.  
No lo invento. Lo leyo.

**Eso es RAG.**

```
SIN RAG:
  Tu pregunta → Modelo → Respuesta inventada

CON RAG:
  Tus datos + Tu pregunta → Modelo → Respuesta basada en datos reales
```

El modelo no cambio. Sus capacidades no cambiaron.  
La diferencia es que ahora tiene el reporte en el escritorio.

---

### Por que no simplemente entrenar el modelo con tus datos?

Buena pregunta. Entrenar un modelo con tus datos cuesta:
- **Tiempo:** semanas o meses de entrenamiento
- **Dinero:** miles o decenas de miles de dolares en GPUs
- **Rigidez:** cuando tus datos cambian, tienes que reentrenar todo

RAG es diferente. No entrenas nada.  
Solo le pasas los datos en el momento de la pregunta.  
Si tus datos cambian manana, RAG funciona igual — solo cambias los datos que le pasas.

Por eso los bancos, hospitales y empresas usan RAG  
para construir asistentes sobre sus propios datos confidenciales.

---


## Concepto 2 — Los 4 bloques de cualquier RAG

---

Todo sistema RAG, sin importar que tan complejo sea,  
tiene exactamente estos 4 bloques.

---

### Bloque 1 — El Contexto (tus datos)

Es la informacion real que le das al modelo antes de cada pregunta.  
Puede ser:
- Una tabla de datos (como el dataset del banco)
- Un documento PDF
- El contenido de una pagina web
- Un correo electronico
- El manual de tu empresa
- Cualquier texto que tenga informacion relevante

**La regla:** cuanto mas relevante y especifico sea el contexto,  
mejores y mas precisas seran las respuestas.

El limite practico es el **tamanio de contexto** del modelo.  
DeepSeek puede recibir hasta ~128,000 tokens en un solo mensaje.  
Un token es aproximadamente 4 caracteres.  
Eso es suficiente para documentos largos, pero no para datasets completos de millones de filas.

Por eso en FraudBot enviamos una muestra de 200 filas, no las 50,000.

---

### Bloque 2 — El System Prompt (las instrucciones)

Es el mensaje que le dice al modelo quien es, como debe comportarse  
y que restricciones tiene. Se envia con **cada pregunta**, aunque el usuario no lo vea.

Ejemplos de lo que puede incluir:
- "Eres FraudBot, analista del Banco Nacional RD"
- "NUNCA inventes numeros — usa solo los datos que tienes"
- "Responde siempre en espanol"
- "Maximo 4 parrafos por respuesta"

**La diferencia entre System Prompt y Contexto:**
- El **contexto** son tus datos (numeros, tablas, documentos)
- El **system prompt** son las instrucciones de comportamiento

Juntos forman lo que el modelo "sabe y siente" antes de leer tu pregunta.

---

### Bloque 3 — El Historial (la memoria)

Los modelos de IA no tienen memoria entre llamadas.  
Cada llamada es independiente por defecto.

Si preguntas "cuantos fraudes hay?" y el modelo responde "492",  
y luego preguntas "y cual es su monto promedio?",  
el modelo no sabe a que "ellos" te refieres — no recuerda la respuesta anterior.

**El historial resuelve eso.**  
Es una lista que guarda todos los mensajes anteriores  
y los envia con cada nueva pregunta.

```python
historial = [
    {"role": "user",      "content": "cuantos fraudes hay?"},
    {"role": "assistant", "content": "Hay 492 fraudes..."},
    {"role": "user",      "content": "cual es su monto promedio?"},  # <- sabe a que se refiere
]
```

Cada vez que el usuario hace una pregunta:
1. Se agrega al historial
2. Se envia TODO el historial junto con el system prompt y el contexto
3. La respuesta se agrega al historial tambien

Asi el modelo siempre tiene toda la conversacion disponible.

---

### Bloque 4 — La Funcion (el mecanismo)

Es el codigo que conecta los 3 bloques anteriores con el modelo.  
Recibe la pregunta del usuario, arma el mensaje completo  
y devuelve la respuesta.

```
preguntar("tu pregunta")
    │
    ├── agrega pregunta al historial
    │
    ├── construye mensaje:
    │     [system prompt + contexto]
    │     + [historial completo]
    │     + [pregunta nueva]
    │
    ├── envia a DeepSeek
    │
    └── guarda respuesta en historial → muestra en pantalla
```

Eso es todo. Cuando entiendes estos 4 bloques,  
puedes construir un RAG para cualquier caso de uso.

## Paso 2 de 5 — Bloque 1: Prepara tu contexto

### La decision de diseno — que datos incluir

No puedes meter todo el dataset en el contexto.  
Tienes que elegir que informacion es mas relevante para las preguntas  
que tus usuarios van a hacer.

Para un analista de fraude, lo mas util es:
- Las estadisticas globales (totales, promedios, porcentajes)
- La distribucion por hora (patron temporal del fraude)
- Una muestra representativa de filas reales

Para otro caso de uso — por ejemplo, un asistente de RRHH —  
el contexto seria el manual de empleados, las politicas salariales, etc.

**El principio es el mismo: preparas los datos relevantes  
y los conviertes en texto que el modelo pueda leer.**


````python
import pandas as pd
from openai import OpenAI

# ── Paso 1: Conectar con DeepSeek ────────────────────────────────────
client = OpenAI(
    api_key="sk-23261b5ee48143d38288c5a86ebb156f",
    base_url="https://api.deepseek.com",
)
print("Conexion con DeepSeek: OK")

# ── Paso 2: Cargar los datos ──────────────────────────────────────────
URL = "https://raw.githubusercontent.com/juanmiguelsuero/fraudguard-ai/main/creditcard_mini.csv"
print("Descargando dataset...")
df = pd.read_csv(URL)
print("Dataset cargado: {:,} filas".format(len(df)))

# ── Paso 3: Calcular estadisticas ────────────────────────────────────
total      = len(df)
fraudes    = int(df["Class"].sum())
pct        = round(df["Class"].mean() * 100, 2)
ratio      = int((df["Class"] == 0).sum() // fraudes)
amt_fraud  = round(df[df["Class"] == 1]["Amount"].mean(), 2)
amt_leg    = round(df[df["Class"] == 0]["Amount"].mean(), 2)
fraude_max = round(df[df["Class"] == 1]["Amount"].max(), 2)
fraude_min = round(df[df["Class"] == 1]["Amount"].min(), 2)

# ── Paso 4: Calcular distribucion por hora ────────────────────────────
df["hora"] = (df["Time"] / 3600 % 24).astype(int)
fraudes_hora = (
    df[df["Class"] == 1]
    .groupby("hora").size()
    .reindex(range(24), fill_value=0)
)
hora_pico = int(fraudes_hora.idxmax())
pico_cant = int(fraudes_hora.max())
dist_horas = "\n".join([
    "  {:02d}:00h -> {} fraudes".format(h, int(fraudes_hora[h]))
    for h in range(24)
])

# ── Paso 5: Seleccionar muestra representativa ────────────────────────
# Elegimos columnas informativas — no necesitamos todas las 30 variables
# V14 y V17 tienen el mayor poder predictivo segun SHAP
cols = ["Time", "V14", "V17", "Amount", "Class"]
muestra = (
    pd.concat([
        df[df["Class"] == 1][cols].head(50),   # 50 fraudes reales
        df[df["Class"] == 0][cols].head(150),  # 150 legitimas reales
    ])
    .sample(frac=1, random_state=42)
    .reset_index(drop=True)
)
muestra["hora"] = (muestra["Time"] / 3600 % 24).round(1)
muestra = muestra.drop(columns=["Time"])
muestra_txt = muestra.round(3).to_string(index=True)

# ── Paso 6: Convertir todo a texto (el contexto) ──────────────────────
# ESTO ES EL CONTEXTO — los datos convertidos a texto que el modelo puede leer
MI_CONTEXTO = (
    "ESTADISTICAS DEL DATASET\n"
    "=" * 50 + "\n"
    "- Transacciones totales: {:,}\n".format(total) +
    "- Fraudes: {} ({:.2f}%)\n".format(fraudes, pct) +
    "- Por cada fraude hay {} transacciones legitimas\n".format(ratio) +
    "- Monto promedio fraude: ${}\n".format(amt_fraud) +
    "- Monto promedio legitima: ${}\n".format(amt_leg) +
    "- Fraude mas alto: ${} | Fraude mas bajo: ${}\n".format(fraude_max, fraude_min) +
    "- Hora pico: {:02d}:00h ({} fraudes en esa hora)\n\n".format(hora_pico, pico_cant) +
    "DISTRIBUCION POR HORA\n"
    "=" * 50 + "\n" +
    dist_horas + "\n\n" +
    "MUESTRA REAL DE 200 FILAS (50 fraudes + 150 legitimas)\n"
    "Clave: V14 y V17 muy negativos = senal fuerte de fraude\n"
    "=" * 50 + "\n" +
    muestra_txt
)

# Mostrar cuanto texto generamos
palabras = len(MI_CONTEXTO.split())
print()
print("=" * 50)
print("  CONTEXTO CONSTRUIDO")
print("=" * 50)
print("  Estadisticas globales: OK")
print("  Distribucion horaria (24 horas): OK")
print("  Muestra de 200 filas reales: OK")
print("  Tamanio del contexto: ~{:,} palabras".format(palabras))
print()
print("  Este es el 'expediente' que el modelo va a tener")
print("  en el escritorio antes de responder cada pregunta.")
print("=" * 50)
print()
print("Bloque 1 listo. Continua al Paso 3.")
````

## Paso 3 de 5 — Bloque 2: Define el System Prompt

---

### Que hace este paso

El system prompt le dice al modelo **quien es** y **como debe comportarse**.  
Se envia con cada pregunta, pero el usuario nunca lo ve directamente.

---

### El system prompt es la diferencia entre un chatbot generico y uno especializado

Sin system prompt, DeepSeek es un asistente general que responde de todo.

Con un system prompt bien escrito, se convierte en:
- Un analista de fraude del Banco Nacional RD
- Un asistente medico de un hospital
- Un agente de soporte de tu empresa
- Lo que necesites

---

### Las 3 partes de un buen system prompt

**1. Identidad** — quien es el bot y para quien trabaja  
**2. Reglas de comportamiento** — que puede y no puede hacer  
**3. El contexto** — los datos reales que tiene disponibles

Las partes 1 y 2 las escribes tu.  
La parte 3 es el contexto que preparaste en el Paso 2.

## Paso 4 de 5 — Bloques 3 y 4: El historial y la funcion

---

### Bloque 3: El historial — la memoria del bot

Sin historial, el bot olvida cada respuesta que da.  
Cada pregunta seria completamente independiente.

El historial es simplemente una lista de Python  
donde guardamos cada mensaje — tanto del usuario como del bot.

```python
historial = []   # empieza vacio

# Despues de la primera pregunta queda asi:
historial = [
    {"role": "user",      "content": "cuantos fraudes hay?"},
    {"role": "assistant", "content": "Hay 492 fraudes (0.98%)..."},
]

# Despues de la segunda:
historial = [
    {"role": "user",      "content": "cuantos fraudes hay?"},
    {"role": "assistant", "content": "Hay 492 fraudes (0.98%)..."},
    {"role": "user",      "content": "cual es su monto promedio?"},
    {"role": "assistant", "content": "El monto promedio de fraude es $88..."},
]
```

Cada vez que mandamos una pregunta, enviamos TODO el historial.  
Asi el modelo tiene la conversacion completa y puede hacer referencias.

---

### Bloque 4: La funcion — el mecanismo que conecta todo

La funcion `preguntar()` es donde se juntan los 4 bloques:

```
preguntar("tu pregunta")
    │
    ├─ Bloque 3: agrega tu pregunta al HISTORIAL
    │
    ├─ Construye el mensaje completo:
    │     Bloque 2 (SYSTEM PROMPT + CONTEXTO)
    │     + Bloque 3 (HISTORIAL con todas las preguntas anteriores)
    │     + tu pregunta nueva
    │
    ├─ Envia a DeepSeek
    │
    └─ Bloque 3: guarda la respuesta en el HISTORIAL
       Muestra la respuesta en pantalla
```

Eso es RAG. Cuatro bloques. Una funcion. Un patron que se repite igual  
para chatbots de bancos, hospitales, empresas legales, o cualquier caso.


````python
# ── Bloque 3: Inicializar el historial ───────────────────────────────
# Una lista vacia al principio — se llena con cada pregunta y respuesta
MI_HISTORIAL = []

# ── Bloque 4: La funcion que conecta todo ────────────────────────────
def preguntar(pregunta):
    """
    Envia una pregunta al bot y muestra la respuesta.

    Internamente:
    - Agrega la pregunta al historial
    - Construye el mensaje con system prompt + contexto + historial + pregunta
    - Envia a DeepSeek
    - Guarda la respuesta en el historial
    - Muestra la respuesta formateada
    """

    # Bloque 3: Guardar la pregunta en el historial
    MI_HISTORIAL.append({
        "role": "user",
        "content": pregunta
    })

    # Construir el mensaje completo para DeepSeek:
    # [Bloque 2: system prompt con contexto]
    # + [Bloque 3: todo el historial de la conversacion]
    # La nueva pregunta ya esta en MI_HISTORIAL (la agregamos arriba)
    mensaje_completo = (
        [{"role": "system", "content": MI_SYSTEM_PROMPT}]
        + MI_HISTORIAL
    )

    # Enviar a DeepSeek
    respuesta_api = client.chat.completions.create(
        model="deepseek-chat",
        messages=mensaje_completo,
        max_tokens=600,   # largo maximo de la respuesta
        temperature=0.3,  # 0=exacto/rigido, 1=creativo/variado
                          # 0.3 = preciso con los datos pero con lenguaje natural
    )

    # Extraer el texto de la respuesta
    respuesta_texto = respuesta_api.choices[0].message.content

    # Bloque 3: Guardar la respuesta en el historial
    # (para que el bot la recuerde en la siguiente pregunta)
    MI_HISTORIAL.append({
        "role": "assistant",
        "content": respuesta_texto
    })

    # Mostrar formateado
    print("-" * 60)
    print("Pregunta: " + pregunta)
    print("-" * 60)
    print()
    for linea in respuesta_texto.split("\n"):
        print("  " + linea)
    print()
    return respuesta_texto


def nueva_conversacion():
    """
    Borra el historial — el bot olvida la conversacion anterior.
    El contexto y el system prompt NO se borran.
    """
    global MI_HISTORIAL
    MI_HISTORIAL = []
    print("Historial borrado. El bot empieza desde cero.")
    print("(Los datos del dataset siguen disponibles)")


# Confirmar que todo esta listo
print("=" * 50)
print("  LOS 4 BLOQUES ESTAN LISTOS")
print("=" * 50)
print()
print("  Bloque 1 - Contexto:     Los datos del dataset")
print("  Bloque 2 - System Prompt: Instrucciones del bot")
print("  Bloque 3 - Historial:     Memoria de la conversacion (vacia)")
print("  Bloque 4 - Funcion:       preguntar() conecta todo")
print()
print("  Tu RAG esta completo.")
print("  Continua al Paso 5 para probarlo.")
````
